```python
import pandas as pd
import glob
import json
```

```python
DATA_PATH = "../data/CLEVR_v1.0/"
```

```python
train_annot_list = []
with open(DATA_PATH+'scenes/CLEVR_train_scenes.json') as f:
    data = json.load(f)
    for scene in data["scenes"]:
        temp = []
        annot = f"There are "
        for scene_object in scene["objects"]:
            color = scene_object["color"]
            size = scene_object["size"]
            shape = scene_object["shape"]
            material = scene_object["material"]
            annot += f"{color} {size} {shape} made of {material}, "

        temp.append(DATA_PATH+'images/train/'+scene["image_filename"])
        temp.append(annot[:-2])
        train_annot_list.append(temp)
    
labels=['Path','Annotation']
train_annotation_dataframe = pd.DataFrame.from_records(train_annot_list, columns=labels)#training Dataframe 
```

```python
train_annotation_dataframe
```

```python
val_annot_list = []
with open(DATA_PATH+'scenes/CLEVR_val_scenes.json') as f:
    data = json.load(f)
    for scene in data["scenes"]:
        temp = []
        annot = f"There are "
        for scene_object in scene["objects"]:
            color = scene_object["color"]
            size = scene_object["size"]
            shape = scene_object["shape"]
            material = scene_object["material"]
            annot += f"{color} {size} {shape} made of {material}, "

        temp.append(DATA_PATH+'images/val/'+scene["image_filename"])
        temp.append(annot[:-2])
        val_annot_list.append(temp)
    
labels=['Path','Annotation']
val_annotation_dataframe = pd.DataFrame.from_records(val_annot_list, columns=labels)#training Dataframe 
```

```python
val_annotation_dataframe
```

```python
trainList=[]
with open(DATA_PATH+'questions/CLEVR_train_questions.json') as f:
    data = json.load(f)
    for K in range(len(data['questions'])):
        i = data['questions'][K]
        temp=[]
        for path in glob.glob(DATA_PATH+'images/train/'+i['image_filename']): 
            temp.append(path)
        temp.append(i['question'])
        temp.append(i['answer'])
        trainList.append(temp)
f.close()
labels=['Path','Question','Answer']
train_dataframe = pd.DataFrame.from_records(trainList, columns=labels)#training Dataframe 
```

```python
train_dataframe
```

```python
valList=[]
with open(DATA_PATH+'questions/CLEVR_val_questions.json') as f:
    data = json.load(f)
    for k in range(len(data['questions'])):
        i = data['questions'][k]
        temp=[]
        for path in glob.glob(DATA_PATH+'images/val/'+i['image_filename']): 
            temp.append(path)
        temp.append(i['question'])
        temp.append(i['answer'])
        valList.append(temp)
f.close()
val_dataframe = pd.DataFrame.from_records(valList, columns=labels)#validation Dataframe
```

```python
val_dataframe
```

```python
train_annotation_dataframe.to_csv("train_annotation_dataframe.csv", index=False)
```
